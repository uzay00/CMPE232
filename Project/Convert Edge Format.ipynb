```python
# import Necessary Libraries
import networkx as nx
import pandas as pd
```

```python
# Import Data
G =nx.read_gexf("babel.gexf", relabel= True)
```

```python
nx.write_edgelist(G, "babel.txt")
```

```python
list(G.edges())
```

```python
fromNodes = [u for (u, v) in list(G.edges())]
fromNodes
```

```python
toNodes = [v for (u, v) in list(G.edges())]
toNodes
```

```python
myedges = pd.DataFrame(
    {'fromNodes': fromNodes,
     'toNodes': toNodes
    })
```

```python
myedges
```

```python
myedges.to_csv('babel.txt', sep=',', index=False, header=False)
```
